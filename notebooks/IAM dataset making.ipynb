```python
import glob 
import numpy as np
import pandas as pd
```

```python
with open("rwth splits/test.txt","r") as f:
    train = f.readlines()
```

```python
t = list(map(lambda x: x.replace("\n",""), train))
```

```python
d = []
for i in t:
    folder,sub_folder = i.split("-")[0],i.split("-")[1]
    d.extend(glob.glob("raw/iam/lines/{0}/{0}-{1}/*".format(folder,sub_folder)))
```

```python
len(d)
```

```python
d = [i.split("/")[-1].split(".")[0] for i in d]
```

```python
with open("raw/iam/rwth_split/testset.txt","w") as f:
    f.write("\n".join(d))
```
